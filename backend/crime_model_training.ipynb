```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
import joblib

# Load dataset
df = pd.read_csv("../data/CrimesOnWomenData.csv")

# Clean column names (VERY IMPORTANT)
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Print columns once (for confirmation)
print("Columns in dataset:", df.columns)

# ---- ADJUST THESE IF NEEDED ----
# Expected logical columns:
# City/District  -> city_col
# Crime type     -> crime_col
# Crime count    -> count_col
# Year           -> year_col

city_col  = df.columns[0]      # first column = city/district
crime_col = df.columns[1]      # second column = crime type
year_col  = df.columns[2]      # third column = year
count_col = df.columns[3]      # fourth column = crime count
# --------------------------------

le_city = LabelEncoder()
le_crime = LabelEncoder()

df[city_col] = le_city.fit_transform(df[city_col])
df[crime_col] = le_crime.fit_transform(df[crime_col])

X = df[[city_col, year_col, crime_col]]
y = df[count_col]

model = LinearRegression()
model.fit(X, y)

joblib.dump(model, "../model/crime_model.pkl")

print("✅ Model trained and saved successfully!")
```
